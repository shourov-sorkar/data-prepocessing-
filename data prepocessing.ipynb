##### In the above cell is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable

## Caping the outliers

```python
def detect_outlier(feature):
    Q1 = feature.quantile(0.25)
    Q3 = feature.quantile(0.75)
    
    IQR = Q3-Q1
    
    lower_bound = Q1-(1.5*IQR)
    upper_bound = Q3+(1.5*IQR)
    
    return feature.index[(feature<lower_bound)|(feature>upper_bound)].tolist()
```

```python
for col in num_cols:
    print(col,'-->',detect_outlier(titanic[col]))
```
